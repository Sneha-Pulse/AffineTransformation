# Affine Transformation

Affine transformation in computer vision is a linear mapping method that preserves points, straight lines, and planes. It changes the shape and position of an object or a set of points. It's like moving, stretching, or twisting something on a piece of graph paper.

To implement affine transformation in Python, you typically use image processing libraries such as OpenCV, which provides built-in functions to make such transformations easy to apply.

Includes combinations of the following:

1. Translation (moving)

2. Scaling (resizing)

3. Rotation (spinning)

4. Shearing (skewing)

## Scaling

Enlarging or shrinking an object along its axes (think of zooming in or out on a picture).

**Types:**

Uniform: Scaling equally in all directions.

Non-uniform: Scaling differently along the x and y axes (stretching or squeezing).

```python
# Scaling

# Define the scaling factor to scale the image
scaling_factor = 0.7

# Apply scaling to the image
scaled_image = cv2.resize(image, None, fx=scaling_factor, fy=scaling_factor)
cv2_imshow(scaled_image)
cv2.waitKey(0)
```

## Translation

Moving an object without changing its shape or orientation. It's like sliding a shape across a page.

How it works: Add or subtract a specific amount from all the x and y coordinates of the points representing the object.


```python
# Translation

# Define the scaling factor to reduce the size of the image
scaling_factor_reduce = 0.7

# Apply scaling before translation to reduce the size of the image
scaled_image = cv2.resize(image, None, fx=scaling_factor_reduce, fy=scaling_factor_reduce)
translated_rows, translated_cols = scaled_image.shape[:2]

# Apply the translation
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(scaled_image, translation_matrix, (translated_cols, translated_rows))
cv2_imshow(translated_image)
cv2.waitKey(0)
```

## Shearing

Slanting or skewing an object (imagine a rectangle turning into a parallelogram).

How it works: One axis (x or y) stays fixed while the other axis is shifted proportionally to the distance from the fixed axis.
